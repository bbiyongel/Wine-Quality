```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import csv
f1=open('X_train_std.csv','r',encoding='UTF-8')
f2=open('y_train.csv','r',encoding='UTF-8')

f3=open('X_val_std.csv','r',encoding='UTF-8')
f4=open('y_val.csv','r',encoding='UTF-8')

f5=open('X_test_std.csv','r',encoding='UTF-8')
f6=open('y_test.csv','r',encoding='UTF-8')

X_train_std = []
y_train = []

X_val_std = []
y_val = []

X_test_std = []
y_test = []

for r in csv.reader(f1):
    X_train_std.append(r)
for r in csv.reader(f3):
    X_val_std.append(r)
for r in csv.reader(f5):
    X_test_std.append(r)

for r in csv.reader(f2):
    r=r[0].split(',')
    for i in r:
        y_train.append(i)
for r in csv.reader(f4):
    r=r[0].split(',')
    for i in r:
        y_val.append(i)
for r in csv.reader(f6):
    r=r[0].split(',')
    for i in r:
        y_test.append(i)
        
X_train_std=np.array(X_train_std)
X_train_std = X_train_std.astype(np.float64)
y_train=np.array(y_train)
y_train = y_train.astype(np.float64)

X_val_std=np.array(X_val_std)
X_val_std = X_val_std.astype(np.float64)
y_val=np.array(y_val)
y_val = y_val.astype(np.float64)

X_test_std=np.array(X_test_std)
X_test_std = X_test_std.astype(np.float64)
y_test=np.array(y_test)
y_test = y_test.astype(np.float64)
```

```python
X_train_std.shape
```

# KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc = KNeighborsClassifier(n_neighbors=1)
knc.fit(X_train_std, y_train)
knc.score(X_val_std,y_val)
```

```python
k_list=range(1,150)
accuracies=[]

for k in k_list:
  knc = KNeighborsClassifier(n_neighbors = k, weights='uniform', n_jobs=-1)
  knc.fit(X_train_std, y_train)
  accuracies.append(knc.score(X_val_std, y_val))

plt.plot(k_list, accuracies)
plt.xlabel("k")
plt.ylabel("Validation Accuracy")
plt.show()

print(max(accuracies))
```

```python
k_list=range(1,150)
accuracies=[]

for k in k_list:
  knc = KNeighborsClassifier(n_neighbors = k, weights='distance', n_jobs=-1)
  knc.fit(X_train_std, y_train)
  accuracies.append(knc.score(X_val_std, y_val))

plt.plot(k_list, accuracies)
plt.xlabel("k")
plt.ylabel("Validation Accuracy")
plt.show()

print(max(accuracies), accuracies.index(max(accuracies)))
```
